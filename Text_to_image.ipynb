```python
# Import Required Libraries
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
```

```python
# Load Pretrained Stable Diffusion Model
# Model ID from Hugging Face
model_id = "dreamlike-art/dreamlike-diffusion-1.0"
model_id2 = "stabilityai/stable-diffusion-xl-base-1.0"
# Load the model
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    use_safetensors=True
)

# Move model to GPU
pipe = pipe.to("cuda")
```

```python
# Text prompt
prompt = """
A fantasy landscape with a dragon flying over a medieval castle,
intricate details, highly detailed,digital painting,
artstation, concept art, smooth,
sharp focus, atmospheric,
volumetric lighting
"""

# Generate image
image = pipe(prompt).images[0]
```

```python
# Display Generated Image
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
# Save Generated Image
image.save("fantasy_castle.png")
print("Fantasy image saved successfully.")
```

```python

new_prompt = """
A futuristic cityscape at sunset,
neon lights,
flying cars,
towering skyscrapers,
cyberpunk,
cinematic lighting,
ultra detailed,
8K
"""
new_image = pipe(new_prompt).images[0]
```

```python
# Display Second Image
plt.imshow(new_image)
plt.axis("off")
plt.show()
```

```python
# Image generation parameters
params = {
    "num_inference_steps": 100,
    "width": 512,
    "height": 512,
    "guidance_scale": 7.5
}
```

```python
# Generate high-quality image
high_quality_image = pipe(
    new_prompt,
    **params
).images[0]
```

```python
# Display High Quality Image
plt.imshow(high_quality_image)
plt.axis("off")
plt.show()
```

```python
#Generate Multiple Images
images = pipe(
    prompt,
    num_inference_steps=50,
    num_images_per_prompt=4
).images

# Display and Save Multiple Images
for i, img in enumerate(images):

    plt.imshow(img)
    plt.axis("off")
    plt.show()

    img.save(f"generated_image_{i+1}.png")

print("Multiple images generated and saved successfully.")
```

```python
# Save Stable Diffusion Model
pipe.save_pretrained("stable-diffusion-model")
print(" Model saved successfully.")
```
